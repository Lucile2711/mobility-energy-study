# Exploration des données de déplacements domicile-travail (CO₂)

## Objectif
Explorer les données issues de data.gouv.fr sur les émissions de CO₂ liées
aux déplacements domicile–travail afin d’identifier des indicateurs utiles
pour l’aide à la décision RH et business.

## Jeux de données
- depl_dom_trav_co2_2021.csv
- depl_dom_trav_co2_2022.csv
